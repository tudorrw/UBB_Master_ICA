```python
import spacy
import nltk
from nltk.stem import PorterStemmer
from nltk.stem import SnowballStemmer
from spacy import displacy
import ro_core_news_sm, en_core_web_sm
# import ro_core_news_lg, en_core_web_lg 
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
nlp_en = en_core_web_sm.load()
nlp_ro = ro_core_news_sm.load()

stemmer_en = PorterStemmer()
stemmer_ro = SnowballStemmer("romanian")

with open("sentences/english_sentences.txt", encoding='utf-8') as file:
    en_sentences = file.readlines()

with open("sentences/romanian_sentences.txt", encoding='utf-8') as file:
    ro_sentences = file.readlines()
```

```python
print("\nTokenization | Part of Speech Tagging | Stemming | Lemmatization\n")
for language, spacy, stemmer, sentences in [("English", nlp_en, stemmer_en, en_sentences), ("Romanian", nlp_ro, stemmer_ro, ro_sentences)]:
    for sentence in sentences:
        doc = spacy(sentence[:-1])
        print(f"{language} sentence: {sentence.strip()}")
        print("{:15} | {:10} | {:10} | {:10} | {:10} | {:10} | {:10}".format('Token','Pos-Tag','Stemming','Lemma', 'Dep','Head','NER'))

        print("-" * 90)
        print("\n".join([f"{token.text:<15} | {token.pos_:<10} | {stemmer.stem(token.text):<10} | {token.lemma_:<10} | {token.dep_:<10} | {token.head.text:<10} | {token.ent_type_:<10}" for token in doc]))
        print("\n")
    print("\n")
```

```python
print("Chunking\n")
grammar = "NP: {<DT>?<JJ>*<NN>}"
cp = nltk.RegexpParser(grammar)

for language, sentences in [("English", en_sentences), ("Romanian", ro_sentences)]:
    for sentence in sentences:
        tokens = nltk.word_tokenize(sentence)
        pos_tags = nltk.pos_tag(tokens)
        
        tree = cp.parse(pos_tags)
        nltk_chunks = [
            " ".join(w for w, _ in subtree.leaves())
            for subtree in tree.subtrees()
            if subtree.label() == "NP"
        ]

        print(f"{language} sentence: {sentence.strip()}")
        print(f"Chunks: {nltk_chunks}")
        print("\n")
```

```python
options = { 
    'bg': '#f0f0f0',
    "font": "Source Sans Pro",
    "bg": "#f0f0f0",
    "arrow_stroke": 2,
    "arrow_width": 8,
    "word_spacing": 30,
    "distance": 100,
    "fine_grained": False,
    "compact": True,
    "font_size": "1.2em"    
}
for language, spacy, sentences in [("English", nlp_en, en_sentences), ("Romanian", nlp_ro, ro_sentences)]:
    for sentence in sentences:
        doc = spacy(sentence[:-1])
        print(f"{language} sentence: {sentence.strip()}")
        displacy.render(doc, style="dep", options=options, jupyter=True)
```

```python
def process_nlp_data(lang_name, nlp_model, stemmer, sentences):
    results = [f"--- {lang_name} RESULTS ---\n\n"]
    html_body = f"<h1>{lang_name} NLP Analysis</h1>"
    
    for i, sent in enumerate(sentences):
        sent = sent.strip()
        doc = nlp_model(sent)
        results.append(f"Sentence {i+1}: {sent}\n")
        header = "{:15} | {:10} | {:15} | {:15} | {:12} | {:10} | {:10}\n".format(
            'Token', 'Pos-Tag', 'Stemming', 'Lemma', 'Dep', 'Head', 'NER'
        )
        results.append(header + "-"*100 + "\n")
        
        for token in doc:
            stem = stemmer.stem(token.text)
            line = "{:15} | {:10} | {:15} | {:15} | {:12} | {:10} | {:10}\n".format(
                token.text, token.pos_, stem, token.lemma_, token.dep_, token.head.text, token.ent_type_
            )
            results.append(line)

        tokens = nltk.word_tokenize(sent)
        pos_tags = nltk.pos_tag(tokens)
        tree = cp.parse(pos_tags)
        
        chunks = [" ".join(w for w, t in s.leaves()) for s in tree.subtrees() if s.label() == 'NP']
        results.append("\nChunks: " + ", ".join(chunks) + "\n")
        results.append("="*100 + "\n\n")

        svg = displacy.render(doc, style="dep", jupyter=False, options={'bg': '#f0f0f0', 'compact': True})
        html_body += f"<h3>Sentence {i+1}: {sent}</h3><div style='background:#f0f0f0; overflow-x:auto;'>{svg}</div><br>"

    return "".join(results), f"<html><body style='font-family:sans-serif;'>{html_body}</body></html>"


en_text, en_html = process_nlp_data("ENGLISH", nlp_en, stemmer_en, en_sentences)
ro_text, ro_html = process_nlp_data("ROMANIAN", nlp_ro, stemmer_ro, ro_sentences)

import os
os.makedirs("results", exist_ok=True)

with open("results/results.txt", "w", encoding="utf-8") as f:
    f.write(en_text + "\n" + ro_text)

with open("results/English_Results.html", "w", encoding="utf-8") as f:
    f.write(en_html)

with open("results/Romanian_Results.html", "w", encoding="utf-8") as f:
    f.write(ro_html)
```
