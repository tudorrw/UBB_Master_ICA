```python
print("Unique values for id_alg:", df["id_alg"].unique())
print("Unique values for id_dataset:", df["id_dataset"].unique())

print("Memory usage for id_alg and id_dataset:", df.id_alg.memory_usage(deep=True), "/  ",  df.id_dataset.memory_usage(deep=True))
df["id_alg"] = df["id_alg"].astype("category")
df["id_dataset"] = df["id_dataset"].astype("category")
print("Memory usage for id_alg and id_dataset after conversion:", df.id_alg.memory_usage(deep=True), "/  ",  df.id_dataset.memory_usage(deep=True))
```

```python
nr_datasets = len(df['id_dataset'].unique())
dataset_ids = sorted(df['id_dataset'].unique())
algorithm_ids = df['id_alg'].unique()
```

```python
# Group by the id of the algorithm and the id of the dataset, and compute the maximum of mean_ind for each group
aggregation_res = df.groupby(["id_alg", "id_dataset"], observed=False).agg(
    max_mean=("mean_ind", "max"), 
    min_mean=("mean_ind", "min"),
    avg_mean=("mean_ind", "mean"),
    max_std=("std_ind", "max"),
    min_std=("std_ind", "min"),
    avg_std=("std_ind", "mean"),
).reset_index()
```

```python
def plot_bar_results(title, y_axis):
    rows, cols = nr_datasets // 4, nr_datasets // 2
    fig, axes = plt.subplots(rows, cols, figsize=(26, 10))
    fig.suptitle(title, fontsize=16)
    for i in range(cols):
        for j in range(rows):
            idx = i * 2 + j
            dataset_results = aggregation_res[aggregation_res["id_dataset"] == dataset_ids[idx]]
            ax = plt.subplot(rows, cols, idx + 1)
            sns.barplot(x="id_alg", y=y_axis, data=dataset_results, hue="id_alg", width=0.8, ax=ax)
            ax.set_title(f"Dataset {dataset_ids[idx]}")
            ax.set_xlabel("Algorithm")
            ax.set_ylabel("Mean Index")
            ax.legend_ = None
            for container in ax.containers:
                ax.bar_label(container, fmt='%.3f', fontsize=8, label_type='edge'),
        
    plt.tight_layout()
    fig.subplots_adjust(wspace=0.2)
```

Mean of the Metric used over the ten Folds

```python
plot_bar_results("Maximum Mean Indices for each model on all datasets", "max_mean")
```

```python
plot_bar_results("Minimum Mean Indices for each model on all datasets", "min_mean")
```

```python
plot_bar_results("Average Mean Indices for each model on all datasets", "avg_mean")
```

the results of the models revolve around the same values for each dataset, but indeed there are slightly small differences. 

Depending on the dataset, the algorithms either perform well (with almost perfect scores, ex: Dataset 6) or poorly (even approaching a random classifier ex: Dataset 8). This diversification shows that none of the algorithm is better in particular than the others.

Standard Deviation of the ten Folds

```python
plot_bar_results("Minimum Standard Deviation for each model on all datasets", "min_std")
```

```python
plot_bar_results("Maximum Standard Deviation for each model on all datasets", "max_std")
```

```python
plot_bar_results("Mean Standard Deviation for each model on all datasets", "avg_std")
```

```python
mean_res = df.groupby('id_alg').agg(
    avg_mean=('mean_ind', 'mean'),
    avg_std=('std_ind', 'mean')
).reset_index()
mean_res
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='avg_mean', y='avg_std', hue='id_alg', data=mean_res,  style='id_alg', s=200, palette='tab10')
plt.title('Stability (std_ind) vs. Performance (mean_ind)')
plt.xlabel('Average Mean Performance (Higher is Better)')
plt.ylabel('Average Standard Deviation (Lower is Better)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

This scatter plot represents a comparison between the mean performance and the mean standard deviation obtained on the datasets for each algorithm. As we can see we are looking for the ones that are situated in the lower-right corner, these are the best performers because they maintain high metric scores while keeping the deviation as low as possible. By looking at the plot we can conclude that Algorithm 8, 10 and 12 would fit the best to the mentioned description.

```python
folds = list(map(lambda x: f'fold_{x}', range(0, 10)))
folds
```

```python
df_long = df.melt(
    id_vars=['id_alg', 'id_dataset'], 
    value_vars=folds, 
    var_name='fold_id', 
    value_name='metric_score'
)
```

```python
df_long.head(50)
```

```python
order = df_long.groupby('id_alg', observed=False)['metric_score'].median().sort_values(ascending=False).index

plt.figure(figsize=(14, 7))
sns.boxplot(
    x='id_alg', 
    y='metric_score', 
    data=df_long, 
    order=order, 
    palette='viridis'
)

plt.title('Performance Distribution Across All K-Folds')
plt.ylabel('Metric Score')
plt.xlabel('Algorithm (Sorted by Median)')
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

This boxplot shows the median line and the Interquartile Range (IQR) alongside their whiskers and a few outliers. 
We are looking for the model that has the highest median and the narrowest box, indicating it is consistent regardless of the fold or dataset.
Algorithm 12 and Algorithm 5 actually show the highest median scores.
